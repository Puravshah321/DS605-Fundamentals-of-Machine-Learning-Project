```python
import pandas as pd

# Load your CSV file
df = pd.read_csv("HI-Small_Trans.csv")  # replace with your actual filename

# Check counts of 1 and 0 in the 'Is Laundering' column
counts = df['Is Laundering'].value_counts()

# Print results
print("Count of 1s (Laundering):", counts.get(1, 0))
print("Count of 0s (Not Laundering):", counts.get(0, 0))

# Optional: print percentage distribution
percentages = df['Is Laundering'].value_counts(normalize=True) * 100
print("\nPercentage distribution:")
print(percentages)
```
